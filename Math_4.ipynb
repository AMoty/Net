```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

### Task 1

```python
def find_der():
    x, y = sy.symbols("x y")
    f = x**5 + 4 * sy.sin(2*x) + sy.cos(3*x +3)
    df = sy.diff(f, x)
    ddf = sy.diff(df, x)
    print(f"Первая производная ({df.subs(x, 1)}) принимает значение {df.evalf(subs={x:1}, n = 4)} в точке х = 1")
    print(f"Вторая производная ({ddf.subs(x, 1)}) принимает значение {ddf.evalf(subs={x:1}, n = 4)} в точке х = 1")
```

```python
find_der()
```

### Task 2

```python
def func1(x):
    x, y = sy.symbols("x y")
    print(sy.diff((sy.sin(2*x + 1)) ** 5))
    print(sy.diff(sy.diff((sy.sin(2*x + 1)) ** 5)))
```

```python
x = np.arange(-5,5,0.1)
```

```python
print(func1(x))
```

```python
y1 = 10*np.sin(2*x + 1)**4*np.cos(2*x + 1)
```

```python
plt.plot(x, y1)
plt.grid()
plt.title("First derivative")
plt.show()
```

```python
y2 = -20*np.sin(2*x + 1)**5 + 80*np.sin(2*x + 1)**3*np.cos(2*x + 1)**2
```

```python
plt.plot(x, y2)
plt.grid()
plt.title("Second derivative")
plt.show()
```

```python
plt.plot(x, y1, label= 'First derivative')
plt.plot(x, y2, label = 'Second derivative')
plt.grid()
plt.legend()
plt.title("Both functions")
plt.show()
```
