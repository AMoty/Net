- Создайте случайную матрицу А1 размерностью 5 на 5
- Транспонируйте матрицу A1
- Найдите обратную матрицу для А1
- Создайте диагональную матрицу - поэлементно перемножив единичную матрицу I и случайную матрицу A2 размерности 4 на 4
- Создайте нулевую матрицу размерности 3 на 3
- Для всех полученных матриц найдите: определитель, ранг и собственные значения
- Проанализируйте результаты

```python
A1 = np.random.random((5,5))
print(A1)
```

```python
print(A1.transpose())
```

```python
from numpy import linalg as ln
A_inv = ln.inv(A1)
print(A_inv)
```

```python
I = np.eye(4) 
A2 = np.random.random((4,4))
A3 = I * A2
print(A3)
```

```python
A4 = np.zeros((3,3))
print(A4)
```

```python
print(f"Определитель случайной матрицы: \t\t{ln.det(A1):.5f}")
print(f"Определитель транспонированной матрицы: \t{ln.det(A1.transpose()):.5f}")
print(f"Определитель обратной матрицы: \t\t{ln.det(A_inv):.5f}")
print(f"Определитель диагональной матрицы: \t\t {ln.det(A3):.5f}")
print(f"Определитель нулевой матрицы: \t\t {ln.det(A4)}")
```

```python
print(f"Ранг случайной матрицы: \t\t{ln.matrix_rank(A1)}")
print(f"Ранг транспонированной матрицы: \t{ln.matrix_rank(A1.transpose())}")
print(f"Ранг обратной матрицы: \t\t{ln.matrix_rank(A_inv)}")
print(f"Ранг диагональной матрицы: \t\t{ln.matrix_rank(A3)}")
print(f"Ранг нулевой матрицы: \t\t{ln.matrix_rank(A4)}")
```

```python
print(f"Собственные значения случайной матрицы: \n{ln.eigvals(A1)}")
print(f"Собственные значения транспонированной матрицы: \n{ln.eigvals(A1.transpose())}")
print(f"Собственные значения обратной матрицы: \n{ln.eigvals(A_inv)}")
print(f"Собственные значения диагональной матрицы: \n{ln.eigvals(A3)}")
print(f"Собственные значения нулевой матрицы: \n{ln.eigvals(A4)}")
```
